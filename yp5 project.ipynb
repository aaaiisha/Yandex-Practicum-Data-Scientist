–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ç–æ, –∫–∞–∫–∞—è —á–∞—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç —Ç–∞—Ä–∏—Ñ , –∞ –∫–∞–∫–∞—è –Ω–µ—Ç

<div class="alert alert-block alert-warning">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</b>

–í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ —ç—Ç–æ –≤—Å–µ-—Ç–∞–∫–∏ –Ω–µ –ª—é–¥–∏ (—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º —É –Ω–∞—Å –ø—Ä–∏–º–µ—Ä–Ω–æ 350 –Ω–∞ —Å–º–∞—Ä—Ç–µ –∏ 150 –Ω–∞ —É–ª—å—Ç—Ä–∞), –∞ —ç—Ç–æ –∏–º–µ–Ω–Ω–æ —Å–∫–æ–ª—å–∫–æ –∑–∞–ø–∏—Å–µ–π –ø–æ –º–µ—Å—è—Ü–∞–º —É –Ω–∞—Å –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞.

</div>



–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —á–∞—Å—Ç–æ—Ç—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è (–ø—Ä–æ–≥–æ–≤–æ—Ä–µ–Ω–Ω—ã–µ –º–∏–Ω—É—Ç—ã, –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–µ –°–ú–° –∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫) –ø–æ –∫–∞–∂–¥–æ–º—É –∏–∑ —Ç–∞—Ä–∏—Ñ–æ–≤:

```python
columns_to_plot = ['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', '–°–ú–°, —à—Ç.', '–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±']

calls_sum = df_event_numbers['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'].sum()
sms_sum = df_event_numbers['–°–ú–°, —à—Ç.'].sum()
internet_sum = df_event_numbers['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'].sum()

def func(pct):
    if pct > 65 or pct < 35:
        sum_of_data = internet_sum
        value = float(pct/100.*sum_of_data/1024)
        metric = '–ì–±'
    elif pct > 64 or pct < 36:
        sum_of_data = calls_sum
        value = float(pct/100.*sum_of_data)
        metric = '–º–∏–Ω.'
    else:
        sum_of_data = sms_sum
        value = float(pct/100.*sum_of_data)
        metric = '—à—Ç.'
    return "{:.2f}%\n({:.0f} {})".format(pct, value, metric)

explode = (0.05, 0)  # —Å–µ–∫—Ç–æ—Ä 'ultra' –Ω–µ–º–Ω–æ–≥–æ –≤—ã–¥–≤–∏–Ω–µ–º

axes = df_event_numbers.groupby('–¢–∞—Ä–∏—Ñ')[columns_to_plot].sum().plot(kind='pie', subplots=True, fontsize=15,
                                                                       use_index=False, shadow=True, explode=explode,
                                                                       title = columns_to_plot, pctdistance=0.45,
                                                                       autopct=lambda pct: func(pct), figsize=(20,5))

for ax in axes:
    ax.legend(['–°–º–∞—Ä—Ç', '–£–ª—å—Ç—Ä–∞'], framealpha=0.2, bbox_to_anchor=(0.87,0.87), prop={'size':15})
    ax.set_ylabel('')
```

```python
def calls_over_plan(row):
    if row['–¢–∞—Ä–∏—Ñ'] == 'smart':
        result = row['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'] - 500
        return np.where(result > 0, result, 0)
    else:
        result = row['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'] - 3000
        return np.where(result > 0, result, 0)
    
    
def sms_over_plan(row):
    if row['–¢–∞—Ä–∏—Ñ'] == 'smart':
        result = row['–°–ú–°, —à—Ç.'] - 50
        return np.where(result > 0, result, 0)
    else:
        result =  row['–°–ú–°, —à—Ç.'] - 1000
        return np.where(result > 0, result, 0)
    

def internet_over_plan(row):
    if row['–¢–∞—Ä–∏—Ñ'] == 'smart':
        result = row['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'] - 15 * 1024
        return np.where(result > 0, result, 0)
    else:
        result = row['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'] - 30 * 1024
        return np.where(result > 0, result, 0)
    
    
def month_revenue(row):
    sum_per_month = 0
    if row['–¢–∞—Ä–∏—Ñ'] == 'smart':
        sum_per_month += np.ceil(row['–ú–∏–Ω—É—Ç—ã —Å–≤–µ—Ä—Ö']) * 3
        sum_per_month += row['–°–ú–° —Å–≤–µ—Ä—Ö'] * 3
        sum_per_month += np.ceil(row['–ú–± —Å–≤–µ—Ä—Ö']/1024) * 200
        return sum_per_month
    
    else:
        sum_per_month += np.ceil(row['–ú–∏–Ω—É—Ç—ã —Å–≤–µ—Ä—Ö']) * 1
        sum_per_month += row['–°–ú–° —Å–≤–µ—Ä—Ö'] * 1
        sum_per_month += np.ceil(row['–ú–± —Å–≤–µ—Ä—Ö']/1024) * 150
        return sum_per_month
```

```python
df_event_numbers['–ú–∏–Ω—É—Ç—ã —Å–≤–µ—Ä—Ö'] = df_event_numbers.apply(calls_over_plan, axis=1)
df_event_numbers['–°–ú–° —Å–≤–µ—Ä—Ö'] = df_event_numbers.apply(sms_over_plan, axis=1)
df_event_numbers['–ú–± —Å–≤–µ—Ä—Ö'] = df_event_numbers.apply(internet_over_plan, axis=1)

df_event_numbers['–ü–ª–∞—Ç–∞ —Å–≤–µ—Ä—Ö –ª–∏–º–∏—Ç–∞, —Ä—É–±.'] = df_event_numbers.apply(month_revenue, axis=1)

for column in ['–ú–∏–Ω—É—Ç—ã —Å–≤–µ—Ä—Ö', '–ú–± —Å–≤–µ—Ä—Ö']:
    df_event_numbers[column] = df_event_numbers[column].astype('float')

for column in ['–°–ú–° —Å–≤–µ—Ä—Ö', '–ü–ª–∞—Ç–∞ —Å–≤–µ—Ä—Ö –ª–∏–º–∏—Ç–∞, —Ä—É–±.']:
    df_event_numbers[column] = df_event_numbers[column].astype('int')
```

```python
filter_list = ['smart', 'ultra']
titles_list = ['–ü–æ –∫–ª–∏–µ–Ω—Ç–∞–º —Ç–∞—Ä–∏—Ñ–∞ "–°–º–∞—Ä—Ç"', '–ü–æ –∫–ª–∏–µ–Ω—Ç–∞–º —Ç–∞—Ä–∏—Ñ–∞ "–£–ª—å—Ç—Ä–∞"']
rows = 1
cols = 2
filter_list_counter = 0

sum_smart = df_event_numbers[df_event_numbers['–¢–∞—Ä–∏—Ñ'] == 'smart'].count()['User_id']
sum_ultra = df_event_numbers[df_event_numbers['–¢–∞—Ä–∏—Ñ'] == 'ultra'].count()['User_id']

annot_labels = ['–í —Ä–∞–º–∫–∞—Ö —Ç–∞—Ä–∏—Ñ–∞', '–î–æ–∑–∞–∫–∞–∑ —É—Å–ª—É–≥']

def func_pct(pct):
    if pct < 75 and pct > 25:
        value = float(pct/100.*sum_smart)
    else:
        value = float(pct/100.*sum_ultra)
    return "{:.2f}%\n({:.0f} —á–µ–ª.)".format(pct, value)

fig, axes = plt.subplots(rows, cols, figsize=(20,10))

explode = (0.05, 0)  # —Å–µ–∫—Ç–æ—Ä 'ultra' –Ω–µ–º–Ω–æ–≥–æ –≤—ã–¥–≤–∏–Ω–µ–º

for r in range(rows):
    for c in range(cols):
        if filter_list[filter_list_counter] == 'smart':
            colors = ['#92c6ff', '#ff5050']
        else:
            colors = ['#97F0AA', '#ff5050']
        curr_filtered_df = df_event_numbers[df_event_numbers['–¢–∞—Ä–∏—Ñ'] == filter_list[filter_list_counter]].groupby('–ù–µ —Ö–≤–∞—Ç–∏–ª–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞')['User_id']
        ax = curr_filtered_df.count().plot(kind='pie', fontsize=14, use_index=False, shadow=True, explode=explode,
                                           title=titles_list[filter_list_counter], pctdistance=0.55, figsize=(20,5),
                                           autopct=lambda pct: func_pct(pct), startangle=200, ax=axes[c], colors=colors,
                                           labels=None)

        filter_list_counter += 1

plt.tight_layout(pad=3)

for ax in axes:
    ax.legend(['–í —Ä–∞–º–∫–∞—Ö —Ç–∞—Ä–∏—Ñ–∞', '–î–æ–∑–∞–∫–∞–∑ —É—Å–ª—É–≥'], bbox_to_anchor=(0.897,0.87), prop={'size':15})
    ax.set_ylabel('')
    ax.title.set_size(16)
    ax.title.set_position([0.5, 1.15])
```

–î–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞—Å—á–∏—Ç–∞—é —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã—Ö –º–∏–Ω—É—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Å—è—Ü–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞ –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞. –¢–∞–∫–∂–µ –ø–æ—Å—á–∏—Ç–∞—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞. –ü–æ—Å–ª–µ —á–µ–≥–æ –≤–∏–∑—É–∞–ª—å–Ω–æ —Å—Ä–∞–≤–Ω—é –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ, –≥–¥–µ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __smart__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —Ñ–∏–æ–ª–µ—Ç–æ–≤—ã–π —Ü–≤–µ—Ç, –∞ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __ultra__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —Å–∏–Ω–∏–π —Ü–≤–µ—Ç.

```python
smart_data_calls = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.':'mean'}).reset_index()
ultra_data_calls = df_event_numbers.query('–¢–∞—Ä–∏—Ñ== "ultra"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.':'mean'}).reset_index()
#–ì—Ä–∞—Ñ–∏–∫
ax = (
    smart_data_calls
    .sort_values(by='–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', ascending=False)
    .plot(y='–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', x='–ú–µ—Å—è—Ü', kind='bar', title='–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –≤ –º–µ—Å—è—Ü', figsize=(14, 7), alpha=0.3, color='r', grid=True, legend=True,label='–ó–≤–æ–Ω–∫–∏ –°–º–∞—Ä—Ç')
)
(
    ultra_data_calls
    .sort_values(by='–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', ascending=False)
    .plot(y='–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', x='–ú–µ—Å—è—Ü', ax=ax, kind='bar', color='b', alpha=0.2, legend=True,label='–£–ª—å—Ç—Ä–∞ –ó–≤–æ–Ω–∫–∏')
)
plt.xticks(rotation=0)
plt.xlabel('–ú–µ—Å—è—Ü')
plt.ylabel('–ú–∏–Ω—É—Ç—ã')
plt.show() # —Ä–∏—Å—É—é –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É

smart_mean = int(smart_data_calls['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –º–∏–Ω—É—Ç –¥–ª—è "smart"
ultra_mean = int(ultra_data_calls['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –º–∏–Ω—É—Ç –¥–ª—è "ultra"

# –ü–æ–¥—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞
print('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, smart_mean, '–º–∏–Ω—É—Ç', ClrDef)
print('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, ultra_mean, '–º–∏–Ω—É—Ç', ClrDef)
print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, smart_mean - ultra_mean, '–º–∏–Ω—É—Ç', ClrDef)
print()

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"')['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "smart"
smart_var = np.var(smart, ddof=1)  # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_var), ClrDef)

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "ultra"')['–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "ultra"
ultra_var = np.var(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, '{:.2f}'.format(ultra_var), ClrDef)
print('')

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart_std = np.std(smart, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_std), ClrDef)

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra_std = np.std(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è "ultra":', ClrG, '{:.2f}'.format(ultra_std), ClrDef)

print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, '{:.2f}'.format(100 - (smart_std / ultra_std * 100)), '%')
```

```python
smart_data_messages = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–°–ú–°, —à—Ç.':'mean'}).reset_index()
ultra_data_messages = df_event_numbers.query('–¢–∞—Ä–∏—Ñ== "ultra"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–°–ú–°, —à—Ç.':'mean'}).reset_index()

#–ì—Ä–∞—Ñ–∏–∫

ax = (
    smart_data_messages
    .sort_values(by='–°–ú–°, —à—Ç.', ascending=False)
    .plot(y='–°–ú–°, —à—Ç.', x='–ú–µ—Å—è—Ü', kind='bar', title='–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–º—Å –≤ –º–µ—Å—è—Ü', figsize=(14, 7), alpha=0.3, color='green', grid=True, legend=True,label='–°–ú–° –°–º–∞—Ä—Ç')
)
# –†–∏—Å—É—é –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra" - —Å–∏–Ω–∏–π —Ü–≤–µ—Ç
(
    ultra_data_messages
    .sort_values(by='–°–ú–°, —à—Ç.', ascending=False)
    .plot(y='–°–ú–°, —à—Ç.', x='–ú–µ—Å—è—Ü', ax=ax, kind='bar', color='blue', alpha=0.25, legend=True,label='–°–ú–° –£–ª—å—Ç—Ä–∞')
)
plt.xticks(rotation=0)
plt.xlabel('–ú–µ—Å—è—Ü')
plt.ylabel('–°–ú–°')
plt.show() # —Ä–∏—Å—É—é –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É

smart_mean = int(smart_data_messages['–°–ú–°, —à—Ç.'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–º—Å –¥–ª—è "smart"
ultra_mean = int(ultra_data_messages['–°–ú–°, —à—Ç.'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–º—Å –¥–ª—è "ultra"

# –ü–æ–¥—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞
print('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–º—Å –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, smart_mean, '—Å–º—Å', ClrDef)
print('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–º—Å –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, ultra_mean, '—Å–º—Å', ClrDef)
print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, smart_mean - ultra_mean, '—Å–º—Å', ClrDef)
print()

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"')['–°–ú–°, —à—Ç.'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "smart"
smart_var = np.var(smart, ddof=1)  # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_var), ClrDef)

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "ultra"')['–°–ú–°, —à—Ç.'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "ultra"
ultra_var = np.var(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, '{:.2f}'.format(ultra_var), ClrDef)
print('')

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart_std = np.std(smart, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_std), ClrDef)

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra_std = np.std(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è "ultra":', ClrG, '{:.2f}'.format(ultra_std), ClrDef)

print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, '{:.2f}'.format(100-(smart_std / ultra_std * 100)), '%')

```

–î–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞—Å—á–∏—Ç–∞—é —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞ –≤ –∫–∞–∂–¥–æ–º –º–µ—Å—è—Ü–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞. –¢–∞–∫–∂–µ –ø–æ—Å—á–∏—Ç–∞—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞. –ü–æ—Å–ª–µ —á–µ–≥–æ –≤–∏–∑—É–∞–ª—å–Ω–æ —Å—Ä–∞–≤–Ω—é –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ, –≥–¥–µ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __smart__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å __–∑–µ–ª—ë–Ω—ã–π —Ü–≤–µ—Ç__, –∞ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __ultra__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å __–æ—Ä–∞–Ω–∂–µ–≤—ã–π —Ü–≤–µ—Ç__.

```python
# –°–æ–∑–¥–∞—é df –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ—Ç —Å—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Å—è—Ü–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart_data_internet =  df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±':'mean'}).reset_index()

# –°–æ–∑–¥–∞—é df –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ—Ç —Å—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Å—è—Ü–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra_data_internet =  df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "ultra"').groupby(['–¢–∞—Ä–∏—Ñ','–ú–µ—Å—è—Ü']).agg({'–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±':'mean'}).reset_index()

# –†–∏—Å—É—é –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart" - –∑–µ–ª—ë–Ω—ã–π —Ü–≤–µ—Ç
ax = (
    smart_data_internet
    .sort_values(by='–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±', ascending=False)
    .plot(y='–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±', x='–ú–µ—Å—è—Ü', kind='bar', title='–°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –≤ –º–µ—Å—è—Ü (–≤ –º–±)', figsize=(14, 7), alpha=0.5, color='lime', grid=True, legend=True,label = 'Internet smart')
)
# –†–∏—Å—É—é –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra" - –æ—Ä–∞–Ω–∂–µ–≤—ã–π —Ü–≤–µ—Ç
(
    ultra_data_internet
    .sort_values(by='–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±', ascending=False)
    .plot(y='–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±', x='–ú–µ—Å—è—Ü', ax=ax, kind='bar', color='orange', alpha=0.4, legend=True,label = 'Internet ultra')
)
plt.xticks(rotation=0)
plt.xlabel('–ú–µ—Å—è—Ü')
plt.ylabel('–¢—Ä–∞—Ñ–∏–∫')
plt.show() # —Ä–∏—Å—É—é –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É

smart_mean = int(smart_data_internet['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –æ–±—ä—ë–º–∞ —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è "smart"
ultra_mean = int(ultra_data_internet['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'].mean()) # —Ä–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –æ–±—ä—ë–º–∞ —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è "ultra"

# –ü–æ–¥—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞
print('–°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, smart_mean, '–º–±', ClrDef)
print('–°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, ultra_mean, '–º–±', ClrDef)
print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, smart_mean - ultra_mean, '–º–±', ClrDef)
print()

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart = df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "smart"')['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "smart"
smart_var = np.var(smart, ddof=1)  # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_var), ClrDef)

# –†–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra =  df_event_numbers.query('–¢–∞—Ä–∏—Ñ == "ultra"')['–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'] # df —Å —Ç–∞—Ä–∏—Ñ–æ–º "ultra"
ultra_var = np.var(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏
print('–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, '{:.2f}'.format(ultra_var), ClrDef)
print('')

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart"
smart_std = np.std(smart, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_std), ClrDef)

# –†–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra"
ultra_std = np.std(ultra, ddof=1) # —Ä–∞—Å—á—ë—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–ª—è "ultra":', ClrG, '{:.2f}'.format(ultra_std), ClrDef)

print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, '{:.2f}'.format(100 - (smart_std / ultra_std * 100)), '%')
```

–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–≤ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, —Å–ª–µ–¥—É–µ—Ç –æ—Ç–º–µ—Ç–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤–∞–∂–Ω—ã–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏:
* –û–±—ä—ë–º –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __ultra__ –≤ –∫–∞–∂–¥–æ–º –º–µ—Å—è—Ü–µ –±–æ–ª—å—à–µ —á–µ–º –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __smart__
* –°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __ultra__ –Ω–∞ __3.2 –≥–± –±–æ–ª—å—à–µ__ —á–µ–º –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __smart__
* –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __ultra__ –ø–æ—á—Ç–∏ –Ω–∞ __42% –±–æ–ª—å—à–µ__ —á–µ–º –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ __smart__
* –ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç —Ç—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –æ–±–æ–∏—Ö —Ç–∞—Ä–∏—Ñ–æ–≤ –±—ã–ª–æ –∏–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–æ –≤ __–¥–µ–∫–∞–±—Ä–µ__, –∞ –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ –≤ __—è–Ω–≤–∞—Ä–µ__ 

–î–∞–Ω–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –≥–æ–≤–æ—Ä—è—Ç –æ —Ç–æ–º, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å —Ç–∞—Ä–∏—Ñ–æ–º __ultra__ —Ä–∞—Å—Ö–æ–¥—É—é—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –Ω–∞ __18% –±–æ–ª—å—à–µ__ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞, —á–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å —Ç–∞—Ä–∏—Ñ–æ–º __smart__.

###### 8.4 –ê–Ω–∞–ª–∏–∑ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è "–°—É–º–º–∞—Ä–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –ø–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É –≤ –º–µ—Å—è—Ü"
–î–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å 2 –Ω–æ–≤—ã—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞.

–ü–µ—Ä–≤—ã–π - smart_revenue - –æ–Ω –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–ª–∏–µ–Ω—Ç–∞—Ö –ø–æ–ª—å–∑—É—é—â–∏—Ö—Å—è —Ç–∞—Ä–∏—Ñ–Ω—ã–º –ø–ª–∞–Ω–æ–º "smart".
–í—Ç–æ—Ä–æ–π - ultra_revenue - –æ–Ω –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–ª–∏–µ–Ω—Ç–∞—Ö –ø–æ–ª—å–∑—É—é—â–∏—Ö—Å—è —Ç–∞—Ä–∏—Ñ–Ω—ã–º –ø–ª–∞–Ω–æ–º "ultra".
–ü–æ—Å–ª–µ —á–µ–≥–æ —Ä–∞—Å—Å—á–∏—Ç–∞—é —Å—É–º–∞—Ä–Ω—É—é –≤—ã—Ä—É—á–∫—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–ª–∞–Ω–∞ –≤ –º–µ—Å—è—Ü –∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É—é –Ω–∞ —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ, –≥–¥–µ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __smart__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å __—Å–µ—Ä—ã–π —Ü–≤–µ—Ç__, –∞ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É __ultra__ –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å __—Ñ–∏–æ–ª–µ—Ç–æ–≤—ã–π —Ü–≤–µ—Ç__.

```python
# –í—ã—Ä—É—á–∫–∞ –ø–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É "Ultra"
ultra_revenue = df_event_numbers.query('–¢–∞—Ä–∏—Ñ =="ultra"').groupby(by='–ú–µ—Å—è—Ü').agg({'–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.':'sum'}).reset_index()

# –í—ã—Ä—É—á–∫–∞ –ø–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É "Smart"
smart_revenue =df_event_numbers.query('–¢–∞—Ä–∏—Ñ =="smart"').groupby(by='–ú–µ—Å—è—Ü').agg({'–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.':'sum'}).reset_index()

# –†–∏—Å—É—é –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra" - —Ñ–∏–æ–ª–µ—Ç–æ–≤—ã–π —Ü–≤–µ—Ç
ax = (
    ultra_revenue
    .sort_values(by='–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.', ascending=False)
    .plot(y='–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.', x='–ú–µ—Å—è—Ü', kind='bar', title='–í—ã—Ä—É—á–∫–∞ –ø–æ —Ç–∞—Ä–∏—Ñ–Ω–æ–º—É –ø–ª–∞–Ω—É', figsize=(14, 7), alpha=0.4, color='purple', grid=True, legend=True,label='–í—ã—Ä—É—á–∫–∞ –£–ª—å—Ç—Ä–∞')
)
# –†–∏—Å—É—é –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart" - —Å–µ—Ä—ã–π —Ü–≤–µ—Ç
(
    smart_revenue
    .sort_values(by='–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.', ascending=False)
    .plot(y='–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.', x='–ú–µ—Å—è—Ü', ax=ax, kind='bar', color='gray', alpha=0.4, legend=True,label='–í—ã—Ä—É—á–∫–∞ –°–º–∞—Ä—Ç')
)
plt.xticks(rotation=0)
plt.xlabel('–ú–µ—Å—è—Ü')
plt.ylabel('–í—ã—Ä—É—á–∫–∞')
plt.show() # —Ä–∏—Å—É—é –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É

# –ü–æ–¥—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∞—Ä–∏—Ñ–∞
print('–°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º –µ–∂–µ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "smart":', ClrG, '{:.2f}'.format(smart_revenue['–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.'].mean()), '—Ä—É–±–ª–µ–π', ClrDef)
print('–°—Ä–µ–¥–Ω–∏–π –æ–±—ä—ë–º –µ–∂–µ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –¥–ª—è —Ç–∞—Ä–∏—Ñ–∞ "ultra":', ClrG, '{:.2f}'.format(ultra_revenue['–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.'].mean()), '—Ä—É–±–ª–µ–π', ClrDef)
print('–†–∞–∑–Ω–∏—Ü–∞:', ClrG, '{:.1f}'.format((smart_revenue['–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.'].mean() / ultra_revenue['–°—É–º–º–∞ –∑–∞ –º–µ—Å—è—Ü, —Ä—É–±.'].mean() * 100) - 100), '%')
```

–ï—â–µ —Ä–∞–∑ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –Ω–∞—à—É —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∏ –Ω–∞—Ä–∏—Å—É–µ–º –≥–∏—Å—Ç–æ—Ä–≥–∞–º–º—ã

–ù–∞–π–¥–µ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã

–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞—Å—Å—á–∏—Ç–∞–µ—Ç –Ω–∞–º —ç—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã:

```python
def get_intervals(column, data_filter):
    
    if data_filter == '–¢–∞—Ä–∏—Ñ':
        format_name = column
        data1, data2 = 'smart', 'ultra'
        format_end1 = '—Ç–∞—Ä–∏—Ñ–∞ "–°–º–∞—Ä—Ç"'
        format_end2 = '—Ç–∞—Ä–∏—Ñ–∞ "–£–ª—å—Ç—Ä–∞"'
    elif data_filter == 'is_Moscow':
        format_name = 'is_Moscow'
        data1, data2 = 0, 1
        format_end1 = '–¥–ª—è —Ä–µ–≥–∏–æ–Ω–æ–≤'
        format_end2 = '–¥–ª—è –ú–æ—Å–∫–≤—ã'
    
    q1_1 = df_event_numbers[df_event_numbers[data_filter] == data1][column].quantile(0.25)
    q3_1 = df_event_numbers[df_event_numbers[data_filter] == data1][column].quantile(0.75)
    diff_1 = q3_1 - q1_1
    
    if data_filter == 'is_Moscow':
        left_int_1 = np.where( (q1_1 - 1.5 * diff_1) > 550, q1_1 - 1.5 * diff_1, 550)
    else:
        left_int_1 = np.where( (q1_1 - 1.5 * diff_1) > 0, q1_1 - 1.5 * diff_1, 0)
    
    right_int_1 = np.where( (q3_1 + 1.5 * diff_1) > 0, q3_1 + 1.5 * diff_1, 0)
    
    q1_2 = df_event_numbers[df_event_numbers[data_filter] == data2][column].quantile(0.25)
    q3_2 = df_event_numbers[df_event_numbers[data_filter] == data2][column].quantile(0.75)
    diff_2 = q3_2 - q1_2
    
    if data_filter == 'is_Moscow':
        left_int_2 = np.where( (q1_2 - 1.5 * diff_2) > 550, q1_2 - 1.5 * diff_2, 550)
    else:
        left_int_2 = np.where( (q1_2 - 1.5 * diff_2) > 0, q1_2 - 1.5 * diff_2, 0)
        
    right_int_2 = np.where( (q3_2 + 1.5 * diff_2) > 0, q3_2 + 1.5 * diff_2, 0)
    
    print('–ì—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ "{}" {}: {} - {}'.format(column, format_end1, left_int_1, right_int_1))
    print('–ì—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ "{}" {}: {} - {}'.format(column, format_end2, left_int_2, right_int_2))
    
    return (left_int_1, right_int_1), (left_int_2, right_int_2)
```

–ù–∞–π–¥–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–≤–æ–Ω–∫–æ–≤:

```python
calls_int_smart, calls_int_ultra = get_intervals('–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', '–¢–∞—Ä–∏—Ñ')
```

–ù–∞–π–¥–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –°–ú–°:

```python
sms_int_smart, sms_int_ultra = get_intervals('–°–ú–°, —à—Ç.', '–¢–∞—Ä–∏—Ñ')
```

–ù–∞–π–¥–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–æ–≥–æ —Ç—Ä–∞—Ñ–∏–∫–∞:

```python
internet_int_smart, internet_int_ultra = get_intervals('–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±', '–¢–∞—Ä–∏—Ñ')
```

–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –ø–æ —Ç–∞—Ä–∏—Ñ–∞–º –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–≤–æ–Ω–∫–æ–≤, –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–º—Å, –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞. –ü–æ–º–∏–º–æ —ç—Ç–æ–≥–æ, –≤—ã–≤–µ–¥–µ–º –Ω–∞ —Ä–∞—Å—á–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É, –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—é:
–ù–∞–ø–∏—à–µ–º –¥–ª—è —ç—Ç–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏—é:

```python
def plot_hist_comparison(data):
    
    ax = plt.gca()
    func_dict = {'–ó–≤–æ–Ω–∫–∏': '–ó–≤–æ–Ω–∫–∏, –º–∏–Ω.', '–°–ú–°': '–°–ú–°, —à—Ç.', '–ò–Ω—Ç–µ—Ä–Ω–µ—Ç': '–ò–Ω—Ç–µ—Ä–Ω–µ—Ç, –ú–±'}
    xlims = [(calls_int_smart[0], calls_int_smart[1]), (calls_int_ultra[0], calls_int_ultra[1]),
             (sms_int_smart[0], sms_int_smart[1]), (sms_int_ultra[0], sms_int_ultra[1]),
             (internet_int_smart[0], internet_int_smart[1]), (internet_int_ultra[0], internet_int_ultra[1])]
    
    filt_column = func_dict[data]

    # –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    if data == '–ó–≤–æ–Ω–∫–∏':
        title = '–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–≤–æ–Ω–∫–æ–≤, –º–∏–Ω.'
        xlim1 = xlims[0]
        xlim2 = xlims[1]
        text_x = (58, 8)
        text_y = 118
        count_text_x = 1150
        count_text_y = 75
        alpha = 0.8
        # annot_xy = (990, 60)
        # annot_xytext = (970, 70)
    elif data == '–°–ú–°':
        title = '–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –°–ú–°, —à—Ç.'
        xlim1 = xlims[2]
        xlim2 = xlims[3]
        text_x = (5.5, 1)
        text_y = 200
        count_text_x = 148
        count_text_y = 230
        alpha = 1
        # annot_xy = (133, 220)
        # annot_xytext = (130, 260)
    else:
        title = '–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞, –ú–±'
        xlim1 = xlims[4]
        xlim2 = xlims[5]
        text_x = (3300, 200)
        text_y = 130
        count_text_x = 38000
        count_text_y = 82
        alpha = 0.8
        # annot_xy = (31100, 73)
        # annot_xytext = (30400, 86)
    
    # –ø–æ–ª—É—á–∞–µ–º –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–µ –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã
    filtered_smart_data = df_event_numbers[(df_event_numbers['–¢–∞—Ä–∏—Ñ'] == 'smart') &
                                           (df_event_numbers[filt_column] >= xlim1[0]) &
                                           (df_event_numbers[filt_column] <= xlim1[1])][filt_column]
    filtered_ultra_data = df_event_numbers[(df_event_numbers['–¢–∞—Ä–∏—Ñ'] == 'ultra') &
                                           (df_event_numbers[filt_column] >= xlim2[0]) &
                                           (df_event_numbers[filt_column] <= xlim2[1])][filt_column]
    # —Ä–∏—Å—É–µ–º
    filtered_smart_data.plot(kind='hist', bins=40, figsize=(12,6), ec='black', ax=ax, color='#92c6ff', 
                             title=title, xlim=xlim1)
    filtered_ultra_data.plot(kind='hist', bins=70, figsize=(12,6), ec='black', ax=ax, color='#97F0AA', 
                             title=title, xlim=xlim2, alpha=alpha)
    
    # —Ä–∞—Å—á–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    mean_smart = filtered_smart_data.mean()
    mean_ultra = filtered_ultra_data.mean()
    dispersion_smart = filtered_smart_data.var()
    dispersion_ultra = filtered_ultra_data.var()
    std_smart = filtered_smart_data.std()
    std_ultra = filtered_ultra_data.std()
    
    if data == '–°–ú–°':
        smart_data = '–°—Ä–µ–¥–Ω–µ–µ - {:.2f}\n–î–∏—Å–ø–µ—Ä—Å–∏—è - {:.2f}\nstd - {:.2f}'.format(mean_smart, dispersion_smart, std_smart)
        ultra_data = '–°—Ä–µ–¥–Ω–µ–µ - {:.2f}\n–î–∏—Å–ø–µ—Ä—Å–∏—è - {:.2f}\nstd - {:.2f}'.format(mean_ultra, dispersion_ultra, std_ultra)
    else:
        smart_data = '–°—Ä–µ–¥–Ω–µ–µ - {:.0f}\n–î–∏—Å–ø–µ—Ä—Å–∏—è - {:.0f}\nstd - {:.0f}'.format(mean_smart, dispersion_smart, std_smart)
        ultra_data = '–°—Ä–µ–¥–Ω–µ–µ - {:.0f}\n–î–∏—Å–ø–µ—Ä—Å–∏—è - {:.0f}\nstd - {:.0f}'.format(mean_ultra, dispersion_ultra, std_ultra)
    
    legend_second = plt.legend([smart_data, ultra_data], prop={'size':10.5}, 
                                framealpha=0, bbox_to_anchor=(1.0,0.32), loc=4)
    
    
    plt.gca().add_artist(legend_second)
    
    # —Ä–∏—Å—É–µ–º –º–µ–¥–∏–∞–Ω—ã
    plt.axvline(x=filtered_smart_data.median(),linewidth=3, color='#16B4DC', linestyle='--')
    ax.text(filtered_smart_data.median()-text_x[0], text_y, '{0:.0f}'.format(filtered_smart_data.median()), 
            fontweight='bold', fontsize='large')
    plt.axvline(x=filtered_ultra_data.median(),linewidth=3, color='#35B815', linestyle='--')
    ax.text(filtered_ultra_data.median()+text_x[1], text_y, '{0:.0f}'.format(filtered_ultra_data.median()), 
            fontweight='bold', fontsize='large')
    
    plt.legend(['–ú–µ–¥–∏–∞–Ω–∞ –°–º–∞—Ä—Ç', '–ú–µ–¥–∏–∞–Ω–∞ –£–ª—å—Ç—Ä–∞', '–°–º–∞—Ä—Ç', '–£–ª—å—Ç—Ä–∞'], prop={'size':12})
    ax.text(count_text_x, count_text_y, '–†–∞—Å—á–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', fontweight='bold', fontsize='large')
    
    # ax.annotate('–†–∞—Å—á–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', xy=annot_xy, xytext=annot_xytext, arrowprops=dict(facecolor='black', shrink=0.05),
                # horizontalalignment='right', verticalalignment='top')
```

–ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –ø–æ —Ç–∞—Ä–∏—Ñ–∞–º –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–≤–æ–Ω–∫–æ–≤, –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–º—Å, –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞. –ü–æ–º–∏–º–æ —ç—Ç–æ–≥–æ, –≤—ã–≤–µ–¥–µ–º –Ω–∞ —Ä–∞—Å—á–µ—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É, –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—é:
–ù–∞–ø–∏—à–µ–º –¥–ª—è —ç—Ç–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏—é:

 –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–≤–æ–Ω–∫–æ–≤:

```python
plot_hist_comparison('–ó–≤–æ–Ω–∫–∏')
```
